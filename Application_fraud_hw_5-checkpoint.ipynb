```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('applications data.csv')
data.head()
```

```python
data.shape
```

```python
categ_cols=['ssn','firstname','lastname','address','zip5','homephone','fraud_label']
```

```python
for col in categ_cols:
    data[col]=data[col].astype("category")
```

```python
data['date']=pd.to_datetime(data['date'],format='%Y%m%d')
data['dob']=pd.to_datetime(data['dob'],format='%Y%m%d')
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
categ_cols=['record','ssn','firstname','lastname','address','zip5','homephone','fraud_label']
num_records=[]
percent_populated=[]
unique_values=[]
most_common=[]
for col in categ_cols:
    num_records.append(data[col].shape[0])
    percent_populated.append((1-data[col].isnull().sum()/data.shape[0])*100)
    unique_values.append(data[col].nunique())
    most_common.append(data[col].value_counts().idxmax())
    
    
    
    
    
```

```python
stats_data={'Column Name':categ_cols,'# of Records':num_records,
           '% Populated':percent_populated,'Unique Values':unique_values,'Most Common Field Value':most_common}
stats_table=pd.DataFrame(data=stats_data)
```

```python
ssn_count=data['ssn'].value_counts().sort_values(ascending=False).reset_index()\
                        .rename(columns={'index':'SSN','ssn':'Count'})[:20]
```

```python
plt.figure(figsize=(10,8))
# ssn_count.sort_values('Count',inplace=True)
plt.bar(ssn_count['SSN'].astype("str"),ssn_count['Count'])
plt.xticks(fontsize=10,rotation = 90)
plt.yticks(fontsize=10)
plt.yscale('log')
plt.xlabel('SSN Number',fontsize=13)
plt.ylabel('Log scale of count',fontsize=13)
plt.grid(color='orange',linestyle='-', linewidth=0.4)
plt.title('Bar chart showing top 20 common ssn values',fontsize=14)
plt.show()
```

```python
fn_count=data['firstname'].value_counts().sort_values(ascending=False).reset_index()\
                        .rename(columns={'index':'firstname','firstname':'Count'})[:25]
plt.figure(figsize=(12,8))
# ssn_count.sort_values('Count',inplace=True)
plt.bar(fn_count['firstname'],fn_count['Count'])
plt.xticks(fontsize=10,rotation = 90)
plt.yticks(fontsize=10)
plt.yscale('log')
plt.xlabel('First Name',fontsize=13)
plt.ylabel('Log scale of count',fontsize=13)
plt.grid(color='orange',linestyle='-', linewidth=0.4)
plt.title('Bar chart showing top 25 common firstnames',fontsize=15)
# plt.grid(True)
plt.show()
```

```python
def plot_dist(col):
    col_count=data[col].astype('str').value_counts().sort_values(ascending=False).reset_index()\
                        .rename(columns={'index':col,col:'Count'})[:25]
    plt.figure(figsize=(12,8))
    plt.bar(col_count[col],col_count['Count'])
    plt.xticks(fontsize=10,rotation = 90)
    plt.yticks(fontsize=10)
    plt.yscale('log')
    plt.xlabel(f'{col}',fontsize=13)
    plt.ylabel('Log scale of count',fontsize=13)
    plt.grid(which='major',color='orange',linestyle='-', linewidth=0.4)
    plt.title(f"Bar chart showing top 25 common {col} codes  ",fontsize=15)
    plt.show()
```

```python
plot_dist("lastname")
```

```python
plot_dist("address")
```

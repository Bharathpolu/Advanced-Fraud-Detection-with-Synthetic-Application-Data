## Spliit the data(Jan-Oct) into 3 folds

```python
from sklearn.model_selection import KFold
x = Jan_Oct.iloc[:,:-3]
y = Jan_Oct.iloc[:,-1]
train_index_list=[]
test_index_list=[]
kfolds = KFold(n_splits = 3, shuffle = True, random_state = 1) 
for train_index, test_index in kfolds.split(x):
    train_index_list.append(train_index)
    test_index_list.append(test_index)
    print("TRAIN:", train_index, "TEST:", test_index)
    print(len(train_index),len(test_index))
```

```python
train_group_1=Jan_Oct[Jan_Oct.record.isin(train_index_list[0])]
train_group_2=Jan_Oct[Jan_Oct.record.isin(train_index_list[1])]
train_group_3=Jan_Oct[Jan_Oct.record.isin(train_index_list[2])]
test_group_1=Jan_Oct[Jan_Oct.record.isin(test_index_list[0])]
test_group_2=Jan_Oct[Jan_Oct.record.isin(test_index_list[1])]
test_group_3=Jan_Oct[Jan_Oct.record.isin(test_index_list[2])]

OOT=data_ori[data_ori['date']>='2016-11-01']
# OOT.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
OOT.head(1)
```

## Create the FDR function

```python
def FDR_func(group_list,model):
    FDR_list=[]

    for group in group_list:

        x_train,y_train = group[0].iloc[:,:-3],group[0].iloc[:,-1]
        x_test = group[1].iloc[:,:-3]
        x_OOT = group[2].iloc[:,:-3]
        
        model.fit(x_train,y_train) 
        group[0]['predicted_fraud']=model.predict(x_train)
        group[1]['predicted_fraud']=model.predict(x_test)
        group[2]['predicted_fraud']=model.predict(x_OOT)

        group_sort_train=group[0][['record','fraud_label','predicted_fraud']].sort_values('predicted_fraud',ascending=False)
        group_sort_test=group[1][['record','fraud_label','predicted_fraud']].sort_values('predicted_fraud',ascending=False)
        group_sort_OOT=group[2][['record','fraud_label','predicted_fraud']].sort_values('predicted_fraud',ascending=False)

        group_totalfraud_train=group_sort_train[group_sort_train['fraud_label']==1].shape[0]
        group_totalfraud_test=group_sort_test[group_sort_test['fraud_label']==1].shape[0]
        group_totalfraud_OOT=group_sort_OOT[group_sort_OOT['fraud_label']==1].shape[0]

        group_3perc_train=group_sort_train.iloc[:int(group_sort_train.shape[0]*0.03),:]
        group_3perc_test=group_sort_test.iloc[:int(group_sort_test.shape[0]*0.03),:]
        group_3perc_OOT=group_sort_OOT.iloc[:int(group_sort_OOT.shape[0]*0.03),:]

        group_3perc_fraud_train=group_3perc_train[group_3perc_train['fraud_label']==1]['fraud_label'].count()
        group_3perc_fraud_test=group_3perc_test[group_3perc_test['fraud_label']==1]['fraud_label'].count()
        group_3perc_fraud_OOT=group_3perc_OOT[group_3perc_OOT['fraud_label']==1]['fraud_label'].count()

        group_FDR_train=group_3perc_fraud_train/group_totalfraud_train
        group_FDR_test=group_3perc_fraud_test/group_totalfraud_test
        group_FDR_OOT=group_3perc_fraud_OOT/group_totalfraud_OOT

        FDR_list.append([group_FDR_train,group_FDR_test,group_FDR_OOT])
        
        group[0].drop(columns=['predicted_fraud'],axis=1,inplace=True)
        group[1].drop(columns=['predicted_fraud'],axis=1,inplace=True)
        group[2].drop(columns=['predicted_fraud'],axis=1,inplace=True)

    avg_train=np.mean([FDR_list[0][0],FDR_list[1][0],FDR_list[2][0]])
    avg_test=np.mean([FDR_list[0][1],FDR_list[1][1],FDR_list[2][1]])
    avg_OOT=np.mean([FDR_list[0][2],FDR_list[1][2],FDR_list[2][2]])
        
    return avg_train,avg_test,avg_OOT
```

# Gradient boosting regressor
https://towardsdatascience.com/machine-learning-part-18-boosting-algorithms-gradient-boosting-in-python-ef5ae6965be4

Next, we construct and fit our model. max_depth refers to the number of leaves of each tree (i.e. 4) whereas n_estimators refers to the total number of trees in the ensemble. As mentioned previously, the learning_rate hyperparameter scales the contribution of each tree. If you set it to a low value, you will need more trees in the ensemble to fit the training set, but the overall variance will be lower.

1. Build a model with parameters 
2. use cross validation split 3 sets of train/test
3. find the FDR on each set and average them
4. try this model with different parameters


```python
from sklearn import ensemble
GBregressor1 = ensemble.GradientBoostingRegressor(
    max_depth=3,
    n_estimators=100,
    learning_rate=0.1
)
GBregressor2 = ensemble.GradientBoostingRegressor(
    max_depth=2,
    n_estimators=100,
    learning_rate=0.1
)

GBregressor3 = ensemble.GradientBoostingRegressor(
    max_depth=1,
    n_estimators=100,
    learning_rate=0.1
)

GBregressor4 = ensemble.GradientBoostingRegressor(
    max_depth=5,
    n_estimators=100,
    learning_rate=0.1
)

GBregressor5 = ensemble.GradientBoostingRegressor(
    max_depth=5,
    n_estimators=200,
    learning_rate=0.1
)

GBregressor6 = ensemble.GradientBoostingRegressor(
    max_depth=5,
    n_estimators=500,
    learning_rate=0.1
)

GBregressor7 = ensemble.GradientBoostingRegressor(
    max_depth=5,
    n_estimators=100,
    learning_rate=0.01
)

GBregressor8 = ensemble.GradientBoostingRegressor(
    max_depth=5,
    n_estimators=200,
    learning_rate=0.01
)

GBregressor9 = ensemble.GradientBoostingRegressor(
    max_depth=5,
    n_estimators=500,
    learning_rate=0.01
)
```

```python
#run the first set of parameter
group_list=[[train_group_1,test_group_1,OOT],[train_group_2,test_group_2,OOT],[train_group_3,test_group_3,OOT]]
model=GBregressor1 
FDR_func(group_list,model)#this return the train, test, OOT avg FDR rate
```
